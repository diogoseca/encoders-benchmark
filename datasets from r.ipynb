```python
import pandas as pd
from tqdm.auto import tqdm

pd.options.display.max_columns = 20
pd.options.display.max_rows = 10
```

# Fetching datasets

https://vincentarelbundock.github.io/Rdatasets/datasets.html

```python
rdatasets = pd.read_html('https://vincentarelbundock.github.io/Rdatasets/datasets.html')[1].set_index('Item')
rdatasets['CSV'] = 'https://vincentarelbundock.github.io/Rdatasets/csv/' + rdatasets['Package'] + '/' + rdatasets.index + '.csv'
rdatasets['Doc'] = 'https://vincentarelbundock.github.io/Rdatasets/doc/' + rdatasets['Package'] + '/' + rdatasets.index + '.html'
rdatasets
```

```python
rdatasets.to_csv('data/rdatasets.csv')
```

```python
datasets_with_categoricals = rdatasets.query('n_character!=0 and (n_numeric!=0 or n_factor!=0)')
datasets_with_categoricals
```

# Saving filtered datasets

```python
datasets_with_categoricals.to_csv('data/datasets_with_categoricals.csv')
```

```python
for name in tqdm(datasets_with_categoricals.index):
    dataset = pd.read_csv(datasets_with_categoricals.loc[name, 'CSV'], index_col=0)
    dataset.reset_index(drop=True, inplace=True) # in python, indices start at 0, not 1
    dataset.to_csv('data/'+name+'.csv')
```
